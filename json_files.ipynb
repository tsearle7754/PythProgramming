```python
# JSON with JavaScript object notation--- based on two structures, a collection of name/value pairs and an ordered list of values
# .dump, .dumps, .load, .loads

# create json file
import json

# data
data = {}
data["name"] = "Bob"
data["address"] = "1014 Orange St"
data["languages"] = ["SQL", "Java", "Fortran"]

# write data into json file
with open("bob.json", "w") as file:
    json.dump(data, file, indent=4)     # indent - pretty print, not necessary
```

```python
# Read from a json file
import json

with open("bob.json", "r") as file:
    data = json.load(file)

# Access data
print(type(data))                       # if you want to know the data type
print(data)
print(data.get("languages", None))
```

```python
# Create a json file

import json

# deserialization (decoding)
def hello_decoder(obj):
    if isinstance(obj, dict):           # are you a dictionary, do you have a class key, is it hello? 
        cl = obj.get("class", None)
        if cl and cl == "Hello":
            return Hello()
    return obj

class Hello:
    def __init__(self):
        self.msg = "Hello"
    def __str__(self):
        return self.msg

class HelloEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, Hello):
            return {
                "class": "Hello",
                "message": obj.__str__()
            }
        return super().default(obj)

data = {}
data["name"] = "Bob"
data["address"] = "1014 Orange St"
data["languages"] = ["SQL", "Java", "Fortran"]

# write data into a json file
with open("bob.json", "r") as file:
    data = json.load(file, object_hook=hello_decoder)
    
# Access data
print(type(data))                       # if you want to know the data type
print(data)
print(data.get("languages", None))
```

```python
# Lecture 16
# Exceptions-
# during runtime errors, exceptions will be wanted- division by zero, open nonexistant file, undefined behavior
# try, except blocks
# can raise a custom exception using keyword "raise"
# else is executed if there's no exception, final will be executed regardless
def try_syntax(numerator, deniminator):
    try:
        print(f'In the try block: {numerator}/{denominator})
        result = numerator / denominator
    except ZeroDvisionError as zde:
        print(zde)
    else:
        print('The result is: ', result)
        return result
    finally:
        print('Exiting')
print(try_syntax(12, 4))
```
