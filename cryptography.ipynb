```python
# hash is a fingerprint of a file or a value
# hash function is a mathematical function that takesn an input and returns a number, a hash
# deterministic- same message produces same hash
# irreversible- can't determine original message from the hash
# collision resistant- hard to find messages that produce the same hash
# fast to compute

import hashlib

msg = "Hello, World!"
b_msg = msg.encode()

# Generate a hash value for a message
hash_value = hashlib.sha256(b_msg)      # everything should be binary
print(hash_value.hexdigest())

print(msg, b_msg)

```
