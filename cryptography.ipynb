```python
# hash is a fingerprint of a file or a value
# hash function is a mathematical function that takesn an input and returns a number, a hash
# deterministic- same message produces same hash (same input produces same output)
# irreversible- can't determine original message from the hash
# collision resistant- hard to find messages that produce the same hash (avoids two different inputs having the same output)
# fast to compute

import hashlib

msg = "Hello, World!"
b_msg = msg.encode()

# Generate a hash value for a message
hash_value = hashlib.sha256(b_msg)      # everything should be binary
print(hash_value.hexdigest())

print(msg, b_msg)

```

```python
def generate_keys():
    # take 2 prime numbers
    p = 3461
    q = 3557

    # calculate the product of 2 primes
    n = p * q

    # Public (e)
    # gcd(n, e) = 1
    e = 17

    # Private (d)
    phi = (p - 1)(q - 1)
    d = pow(e, -1, phi)
    print(d)

    return (e, n), (d, n)   #(public), (private)

def encrypt(msg, public_key):
    e, n = public_key
    return pow(msg, e, n)

def decrypt(msg, private_key):
    d, n = private_key
    return pow(msg, d, n)

msg = 1111
public_key, private_key = generate_keys()
print(f"Public key (e, n), {public_key}")
print(f"Private key (d, n), {private_key}")

ciphertext = encrypt(msg, public_key)
plaintext = decrypt(ciphertext, private_key)

print(f"Original: {msg}, Ciphertext: {ciphertext}, Decrypted: {plaintext}")

# There are built in cryptography functions that help with secure key exchange and generation
```
